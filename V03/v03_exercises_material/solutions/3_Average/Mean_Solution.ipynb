$\bar{x} = \frac{1}{n}\sum_{i=1}^n x_i  =  \frac{1}{n} (x_1 + \cdots + x_n)$

## Implementation

```python
%%writefile mean.py

#!/usr/bin/python3
from mrjob.job import MRJob

class MyJob(MRJob):

    def mapper(self, _, line):
        line = line.strip()
        numbers = line.split()
        numbers = [int(i) for i in numbers]
        yield None, (len(numbers), sum(numbers))

    def reducer(self, _, count_sum_pairs):
        all_counts = 0
        all_sum = 0
        for pair in count_sum_pairs:
            all_counts += pair[0]
            all_sum += pair[1]
        
        yield "Average", all_sum/all_counts
        
if __name__ == '__main__':
    MyJob.run()
```

## Better Approach

This approach computes the average in the mapper. In the reducer, we compute the weighted average

```python
%%writefile mean2.py

#!/usr/bin/python3
from mrjob.job import MRJob

class MyJob(MRJob):

    def mapper(self, _, line):
        line = line.strip()
        numbers = line.split()
        numbers = [int(i) for i in numbers]
        yield None, (len(numbers), sum(numbers)/len(numbers))

    def reducer(self, _, count_averages_pairs):
        pairs = list(count_averages_pairs)
        
        all_counts = sum([pair[0] for pair in pairs])
        weighted = [pair[0]/all_counts * pair[1] for pair in pairs]
        
        yield "Average", sum(weighted)
        
if __name__ == '__main__':
    MyJob.run()
```
